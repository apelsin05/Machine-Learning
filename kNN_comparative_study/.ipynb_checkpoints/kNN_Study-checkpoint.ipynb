```python
# loading data
DATA_PATH="Mental_Health_and_Social_Media_Balance_Dataset.csv"

if not os.path.exists(DATA_PATH):
    raise FileNotFoundEror(f"File not found at {DATA_PATH}. Modify path, or place file in cwd.")

df=pd.read_csv(DATA_PATH)
print("Dataset loaded. Size: ", df.shape)
```

```python
# pre-processing:
# elliminate user id, convert genders and social media platforms in numerical values, 
# (one-hot encoding for socials)
# Happiness_Index - target

from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler

# drop ID's
df=df.drop('User_ID', axis=1)

#  gender -> numeical
le_gender=LabelEncoder()
df['Gender']=le_gender.fit_transform(df['Gender']) # (Male=1, Female=0, Other=2)

# One-hot for socials
df=pd.get_dummies(df,columns=['Social_Media_Platform'],drop_first=True)

# define x=features and y=target
X=df.drop('Happiness_Index(1-10)',axis=1)
Y=df['Happiness_Index(1-10)']




```

```python
# scale numericals
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
# dividing into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    X_scaled, Y, test_size=0.2, random_state=42
) # X dependents - they are all numpy arrays now, Y - series

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
# see the progress
display(pd.DataFrame(X_train, columns=X.columns).head())

display(pd.DataFrame(X_test, columns=X.columns).head())

display(Y_train.head())

display(Y_test.head())
```

```python
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)


X_train.to_csv("X_train_preprocessed.csv", index=False)
X_test.to_csv("X_test_preprocessed.csv", index=False)
pd.Series(Y_train, name='target').to_csv("Y_train.csv", index=False)
pd.Series(Y_test, name='target').to_csv("Y_test.csv", index=False)
print("Fișiere preprocesate salvate: X_train_preprocessed.csv, X_test_preprocessed.csv, Y_train.csv, Y_test.csv")
```
